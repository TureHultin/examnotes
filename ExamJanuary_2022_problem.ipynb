## Random variable generation and transformation

The purpose of this problem is to show that you can implement your own sampler, this will be built in the following three steps:

1. [2p] Implement a Linear Congruential Generator where you tested out a good combination (a large $M$ with $a,b$ satisfying the Hull-Dobell (Thm 6.8)) of parameters. Follow the instructions in the code block.
2. [2p] Using a generator construct random numbers from the uniform $[0,1]$ distribution.
3. [4p] Using a uniform $[0,1]$ random generator, generate samples from 

$$p_0(x) = \frac{\pi}{2}|\sin(2\pi x)|, \quad x \in [0,1] \enspace .$$

Using the **Accept-Reject** sampler (**Algorithm 1** in TFDS notes) with sampling density given by the uniform $[0,1]$ distribution.

```python

def problem2_LCG(size=None, seed = 0):
    """
    A linear congruential generator that generates pseudo random numbers according to size.
    
    Parameters
    -------------
    size : an integer denoting how many samples should be produced
    seed : the starting point of the LCG, i.e. u0 in the notes.
    
    Returns
    -------------
    out : a list of the pseudo random numbers
    """

    a = 3
    b = 5
    m = 96
    x = seed
    
    numbers = []

    for _ in range(size):
        temp = (a*x+b)%m
        numbers.append(temp)
        x = temp
    
    return numbers

print(problem2_LCG(20))
```

```python
import numpy as np

def problem2_uniform(generator=problem2_LCG, period = 96, size=None, seed=0):
    # 1. Call the provided generator to get 'size' number of integers
    # The docstring says generator(size, seed)

    random_integers = generator(size, seed)
    
    # 2. Convert to numpy array for easy division
    random_integers = np.array(random_integers)
    
    # 3. Normalize by dividing by the period
    # Note: Using period (m) as the divisor maps {0,...,m-1} to [0, 1)
    samples = random_integers / 96
    
    # 4. Return as a list (as requested by the docstring)
    return samples.tolist()

print(problem2_uniform(size=10))

# Create a wrapper or ensure the function signature matches
def my_uniform_wrapper(size, seed):
    # This calls your uniform function with the correct internal generator
    return problem2_uniform(generator=problem2_LCG, size=size, seed=seed)
```

```python

def problem2_accept_reject(uniformGenerator=None, size=None, n_iterations=1, seed=0):
    """
    Takes a generator that produces uniform pseudo random [0,1] numbers 
    and produces samples from (pi/2)*abs(sin(x*2*pi)) using an Accept-Reject
    sampler with the uniform distribution as the proposal distribution
    
    Parameters
    -------------
    generator : a function of the type generator(size,seed) that produces uniform pseudo random
    numbers from [0,1]
    seed : the seed to be used in the generator provided
    size : an integer denoting how many samples should be produced
    
    Returns
    --------------
    out : a list of the pseudo random numbers with the specified distribution
    """
    
    # We use a secondary seed to ensure the U values (for acceptance) 
    # are different from the X values (the candidates)
    current_seed = seed
    accepted_samples = []
    rejected = 0

    for i in range(n_iterations):
        # Since we reject samples, we must keep looping until we have 'size' samples
        while len(accepted_samples) < size:
            # 1. Generate candidate samples from the proposal distribution (Uniform [0,1])
            # We generate a batch to be efficient, though we can do 1 by 1
            candidates = uniformGenerator(size=size, seed=current_seed)
            
            # 2. Generate uniform numbers for the acceptance criteria
            # We increment the seed to get a fresh sequence of random numbers
            current_seed += 1
            u_values = uniformGenerator(size=size, seed=current_seed)
            
            for i in range(len(candidates)):
                x_cand = candidates[i]
                u = u_values[i]
                
                # Acceptance condition: u <= f(x) / (M * g(x))
                # f(x) = (pi/2) * abs(sin(2*pi*x))
                # M = (pi/2), g(x) = 1
                # Ratio simplifies to abs(sin(2*pi*x))
                acceptance_ratio = np.abs(np.sin(x_cand * 2 * np.pi))
                
                if u <= acceptance_ratio:
                    accepted_samples.append(x_cand)
                else:
                    rejected += 1
                    
                # If we reached the desired size, break the loop early
                if len(accepted_samples) == size:
                    break
            
            # Increment seed again if the loop continues
            current_seed += 1
        
    print(len(accepted_samples))
    print(rejected)
    return accepted_samples

print(problem2_accept_reject(uniformGenerator=problem2_uniform, size=50, seed=0))
```

For example, if we enumerate the cities according to alphabetical order, the first city `'Aracaju (SE)'` would correspond to $0$. Each row of the file corresponds to one flight, i.e. it has a starting city and an ending city. We model this as a stationary Markov chain, i.e. each user's travel trajectory is a realization of the Markov chain, $X_t$. Here, $X_t$ is the current city the user is at, at step $t$, and $X_{t+1}$ is the city the user travels to at the next time step. This means that to each row in the file there is a corresponding pair $(X_{t},X_{t+1})$. The stationarity assumption gives that for all $t$ there is a transition density $p$ such that $P(X_{t+1} = y | X_t = x) = p(x,y)$ (for all $x,y$). The transition matrix should be `n_cities` x `n_citites` in size.
3. [2p] Use the transition matrix to compute out the stationary distribution.

```python



import pandas as pd
import numpy as np

# 1. Load the dataset
# List of cities in alphabetical order (Aracaju is index 0)
cities = [
    'Aracaju (SE)', 'Belém (PA)', 'Brasília (DF)', 'Campo Grande (MS)', 
    'Curitiba (PR)', 'Florianópolis (SC)', 'Fortaleza (CE)', 'Goiânia (GO)', 
    'Manaus (AM)', 'Natal (RN)', 'Porto Alegre (RS)', 'Recife (PE)', 
    'Rio de Janeiro (RJ)', 'Salvador (BA)', 'São Paulo (SP)'
]

# Generate random flight data
np.random.seed(42)
n_obs = 1000
user_codes = np.random.randint(100, 250, size=20) # 20 unique users

data = {
    'userCode': np.random.choice(user_codes, n_obs),
    'from': np.random.choice(cities, n_obs),
    'to': np.random.choice(cities, n_obs),
    'flightType': np.random.choice(['Economic', 'First Class', 'Premium'], n_obs),
    'price': np.random.uniform(100, 1500, n_obs),
    'time': np.random.uniform(0.5, 6.0, n_obs)
}

df = pd.DataFrame(data)

# Ensure 'from' and 'to' aren't the same for any row (no self-loops for a single flight)
df = df[df['from'] != df['to']].reset_index(drop=True)

# 2. Fill in variables
# We combine 'from' and 'to' to ensure we count every unique city in the system
all_cities = pd.concat([df['from'], df['to']]).unique()
all_cities.sort() # Alphabetical order as requested

number_of_cities = len(all_cities)
number_of_userCodes = df['userCode'].nunique()
number_of_observations = len(df)

print(f"Cities: {number_of_cities}")
print(f"Users: {number_of_userCodes}")
print(f"Observations: {number_of_observations}")



# Map city names to indices 0 to n-1
city_to_idx = {city: i for i, city in enumerate(all_cities)}

# Initialize transition matrix with zeros
transition_matrix = np.zeros((number_of_cities, number_of_cities))

# Fill counts
for _, row in df.iterrows():
    i = city_to_idx[row['from']]
    j = city_to_idx[row['to']]
    transition_matrix[i, j] += 1

# Normalize rows to sum to 1
# We add a small epsilon or check for zero rows to avoid division by zero
row_sums = transition_matrix.sum(axis=1)
transition_matrix = transition_matrix / row_sums[:, np.newaxis]
print(transition_matrix)



# Method: Solving the system (P.T - I)pi = 0
# Or more simply, using the power method or linalg
eigenvalues, eigenvectors = np.linalg.eig(transition_matrix.T)

# Find the eigenvector corresponding to eigenvalue 1
idx = np.argmin(np.abs(eigenvalues - 1.0))
stationary_dist = np.real(eigenvectors[:, idx])

# Normalize so the probabilities sum to 1
stationary_dist = stationary_dist / stationary_dist.sum()

print("Stationary distribution computed:")
print(stationary_dist)




# Compute P^3
P3 = np.linalg.matrix_power(transition_matrix, 3)

# Aracaju (SE) is index 0 (first in alphabetical order)
prob_back_in_aracaju = P3[0, 0]

print(f"Probability of being back in Aracaju after 3 steps: {prob_back_in_aracaju:.4f}")
```
